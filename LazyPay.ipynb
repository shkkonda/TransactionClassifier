```python
import pandas as pd
```

```python
df = pd.read_excel("LazyPay2021_22.xlsx", header=None)
```

```python
df.head()
```

```python
new_df = pd.DataFrame({'Name': df[::4].reset_index()[0], 'Date': df[1::4].reset_index()[0], 'Amount': df[2::4].reset_index()[0], 'Status': df[3::4].reset_index()[0]})
```

```python
new_df.head()
```

```python
new_df.groupby('Status').Amount.agg(sum)
```

```python
useful_values = ['Lazycash', 'Paid', 'Received', 'Refunded', 'Successful']
rename = ['Cashback', 'Expense', 'Cashback', 'Reversed', 'Repaid']
```

```python
newer_df = new_df[new_df.Status.isin(useful_values)].replace(useful_values, rename)
```

```python
newer_df.Name.unique()
```

```python
newer_df[newer_df.Name == 'RAMESH C']
```

```python
mapping = pd.read_excel("LazyPayMapping.xlsx")
```

```python
mapping.head()
```

```python
final_df = pd.merge(newer_df, mapping, how='left',  left_on='Name', right_on='input')
```

```python
final_df.head()
```

```python
final_df.groupby('mapping').Amount.agg(sum)
```

```python
final_df[final_df.mapping == 'Eatout']
```
